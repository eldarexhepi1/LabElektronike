**Eksperimenti i dyte -Qarku diferencues** 


**1. Qëllimi i Eksperimentit**

Qëllimi i këtij eksperimenti është studimi i sjelljes së qarkut diferencues RC ndaj ndryshimit të frekuencës së sinjalit hyrës dhe verifikimi i karakteristikave të tij në regjim frekuencor. Analizohet varësia e tensionit në dalje **Vout** ndaj tensionit hyrës **Vin** për frekuenca të ndryshme duke ndryshuar kapacitetin total të qarkut. 
Gjithashtu përcaktohet frekuenca karakteristike e qarkut dhe krahasohen rezultatet eksperimentale me modelin teorik të qarkut diferencues.

**Baza Teorike**
**2.1 Modeli i Qarkut**

Qarku i përdorur: Qark diferencues RC

Skema bazë:

 Vin ──||───┬── Vout
      C    |
           R
           |
          GND

Ku:

Sinjali hyrës: Vin=11.45V

Sinjali dalës: Vout=3.8V
Rezistenca: R=98.7Ω
Kapacitorët: në paralel

Kapacitetet totale:
| Kapacitorët paralel | C total |
| ------------------- | ------- |
| 2×68 pF             | 136 pF  |
| 2×271 pF            | 542 pF  |
| 2×331 pF            | 662 pF  |
| 2×47 pF             | 94 pF   |
| 2×10 pF             | 20 pF   |
| 2×471 pF            | 942 pF  |

**2.2 Ekuacionet Themelore**

Për qarkun diferencues RC, impedanca e rezistencës është:
$$
Z_R = R
$$
Ndërsa impedanca e kapacitorit:
$$
Z_C = \frac{1}{j\omega C}
$$
ku:
$$
\omega = 2\pi f
\]
$$
Duke përdorur formulën e ndarësit të tensionit:
$$
\frac{V_{out}}{V_{in}} =
\frac{R}{R+\frac{1}{j\omega C}}
$$

Pas thjeshtimit merret funksioni transferues i qarkut diferencues:
$$
H(j\omega)=\frac{V_{out}}{V_{in}}
=\frac{j\omega RC}{1+j\omega RC}
$$
Amplituda e funksionit transferues është:

$$
|H(j\omega)|=
\frac{\omega RC}{\sqrt{1+(\omega RC)^2}}
$$

Amplifikimi eksperimental llogaritet me:
$$
A=\frac{V_{out}}{V_{in}}
$$
Duke zëvendësuar vlerat e matura:
$$
A=\frac{3.8}{11.45}
$$
$$
A=0.332
$$
Amplifikimi në decibel:
$$
A_{dB}=20\log(A)
$$
$$
A_{dB}=20\log(0.332)
$$
$$
A_{dB}\approx -9.57\,dB
$$
Frekuenca karakteristike e qarkut është:
$$
f_c=\frac{1}{2\pi RC}
$$

**Eksperimenti 3-High-Pass filter** 
**1. QËLLIMI I EKSPERIMENTIT**
Qëllimi i eksperimentit është studimi i sjelljes së filtrit RC high-pass në funksion të frekuencës. U analizua tensioni i daljes dhe zhvendosja fazore për frekuenca të ndryshme. Eksperimenti verifikon teorinë e filtrit high-pass dhe përcaktimin e frekuencës së prerjes.

**2. BAZA TEORIKE**

**2.1 Modeli i Qarkut**

Qarku përbëhet nga një kondensator dhe një rezistencë të lidhur në konfigurim high-pass. Tensioni i daljes merret në rezistencë.

**2.2 Ekuacionet Themelore**

Funksioni i transferimit jepet nga:

$$H(jω)=Vout/Vin= jωRC/(1+jωRC)$$

Frekuenca teorike e prerjes:

$$fc=1/(2πRC)$$
Në frekuenca të ulëta sinjali attenuohet, ndërsa në frekuenca të larta kalon pothuajse i pandryshuar.

U realizua montimi praktik i qarkut RC high-pass. Frekuenca e sinjalit hyrës u ndryshua gradualisht ndërsa amplituda hyrëse u mbajt konstante. U mat tensioni në dalje dhe zhvendosja fazore.

```python
# ============================================
# GRAFIKU HIGH-PASS
# Vout vs Frekuenca
# ============================================

import matplotlib.pyplot as plt
import pandas as pd

# ============================================
# TE DHENAT
# ============================================

data = pd.DataFrame({

    "Frekuenca (Hz)": [278, 1970, 2420, 3230, 4000, 10270, 31000],

    "Vout (V)": [0.11, 1.96, 2.78, 2.45, 3.14, 3.32, 3.47]

})

# ============================================
# GRAFIKU
# ============================================

plt.figure(figsize=(8,5))

plt.semilogx(
    
    data["Frekuenca (Hz)"],
    data["Vout (V)"],
    marker='o'

)

plt.xlabel("Frekuenca (Hz)")
plt.ylabel("Vout (V)")

plt.title("Grafiku i Filtrit RC High-Pass")

plt.grid(True)

plt.show()
```
