**2. Baza teorike**

1. Zona e bllokimit (reverse bias)  
   Për tensione negative rryma është pothuajse zero.

2. Zona eksponenciale (forward bias)  
   Kur tensioni pozitiv arrin rreth 0.6 - 0.7V, rryma fillon të rritet shumë shpejt në mënyrë eksponenciale.

$$

**Eksperimenti 2: Polarizimi i Drejtë i Diodës Zener**

**1. Qëllimi i Eksperimentit**

Qëllimi i këtij eksperimenti është studimi i karakteristikës tension–rrymë të diodës Zener gjatë polarizimit të drejtë. Në këtë eksperiment analizohet qarku i përbërë nga burimi i tensionit, rezistenca kufizuese dhe dioda Zener. Verifikohet sjellja e diodës gjatë përcjelljes së rrymës dhe marrëdhënia ndërmjet tensionit hyrës ($V_{in}$), rrymës ($I$) dhe tensionit në dalje ($V_{out}$). Gjithashtu studiohet ndryshimi i rrymës me rritjen e tensionit të aplikuar dhe vërtetohet sjellja jo-lineare karakteristike e diodave gjysmëpërçuese.


**2.Baza Teorike**

**2.1 Modeli i Qarkut**

Qarku përbëhet nga:

- Burimi i tensionit ($V_{in}$)

- Rezistenca kufizuese:

$$
R=1k\Omega
$$

- Dioda Zener në polarizim të drejtë

- Voltmetër dhe ampermetër

Në polarizim të drejtë terminali pozitiv i burimit lidhet me anodën e diodës, ndërsa terminali negativ me katodën. Në këtë gjendje dioda fillon të përcjellë rrymë pas arritjes së tensionit prag.






**2.2 Ekuacionet Themelore**

Ligji i Ohmit:

$$
I=\frac{V}{R}
$$

ku:

- $I$ = rryma (A)

- $V$ = tensioni në rezistencë (V)

- $R$ = rezistenca ($\Omega$)

Në qark:

$$
V_R=V_{in}-V_{out}
$$

Prandaj:

$$
I=
\frac{V_R}{R}
$$

ose:

$$
I=
\frac{V_{in}-V_{out}}
{R}
$$

#### **Eksperimenti nr 4-Ura e diodave**

 **Qëllimi**
Të studiohet funksionimi i urës së diodave dhe të vëzhgohet forma e sinjalit në dalje me anë të oshiloskopit.

**Pajisjet**
- Ura e diodave (4 dioda)
- Gjenerator sinjali
- Oshiloskop
- Breadboard dhe tela lidhës

**Procedura**
U realizua lidhja e urës së diodave dhe në hyrje u aplikua tension alternativ (AC). Forma e sinjalit në dalje u analizua me oshiloskop.



 **Vëzhgime në oshiloskop**
 
Nga pamja e oshiloskopit u vu re se:

- Sinjali në dalje nuk ka pjesë negative.
- Gjysmëperiodat negative janë kthyer në pozitive.
- Në dalje përftohet tension pulsues pozitiv (DC pulsues).
- Forma e valës tregon rektifikim të plotë (Full-Wave Rectification).

**Interpretimi**

Ura e diodave realizon drejtimin e tensionit alternativ duke bërë që rryma të kalojë gjithmonë në të njëjtin drejtim në dalje. Për këtë arsye, sinjali negativ kthehet pozitiv dhe tensioni në dalje mbetet gjithmonë mbi boshtin zero.

Mund të shprehet si:

$$
V_{out} = |V_{in}|
$$

ku:

$$
V_{in} = V_m \sin(\omega t)
$$

dhe në dalje:

$$
V_{out}=|V_m\sin(\omega t)|
$$

**Përfundim**
Eksperimenti tregoi se ura e diodave funksionon si rektifikues i plotë, duke shndërruar tensionin AC në tension DC pulsues. Rezultatet e oshiloskopit konfirmuan funksionimin korrekt të qarkut.


**Eksperimenti 5-Studimi i Urës së Diodave (Bridge Rectifier)**

**Qëllimi i eksperimentit**

Qëllimi i këtij eksperimenti është studimi i funksionimit të urës së diodave dhe analizimi i formës së tensionit në dalje me anë të oshiloskopit. Eksperimenti demonstron procesin e drejtëzimit të tensionit alternativ (AC) në tension njëkahor pulsues (DC).

**Pajisjet e përdorura**

1. Breadboard
2. 4 dioda për ndërtimin e urës së diodave
3. Gjenerator funksionesh (Function Generator)
4. Oshiloskop digital
5. Rezistencë 
6. Tela lidhës



**Parimi teorik**

Ura e diodave përbëhet nga katër dioda të lidhura në konfigurim ure dhe përdoret për realizimin e drejtëzimit të plotë të tensionit alternativ (Full-Wave Rectification).

Kur tensioni hyrës është pozitiv:
dy dioda përcjellin rrymën;dy diodat e tjera bllokojnë.


Kur tensioni hyrës bëhet negativ:
 përcjell çifti tjetër i diodave;
  drejtimi i rrymës në ngarkesë mbetet i njëjtë.


Si rezultat, të dy gjysmëperiodat e sinjalit dalin me polaritet pozitiv.

Sinjali hyrës jepet nga:

$$
V_{in}(t)=V_0\sin(\omega t)
$$

Ndërsa tensioni në dalje pas urës së diodave është:

$$
V_{out}(t)=|V_0\sin(\omega t)|
$$

Në çdo gjysmëperiodë përcjellin dy dioda, prandaj ekziston humbje tensioni rreth:

$$
\Delta V \approx 2 \times 0.7V = 1.4V
$$



**Vëzhgimet**

Nga pamja e oshiloskopit u vu re se:

sinjali në dalje ishte gjithmonë pozitiv;
     pjesa negative e sinusoidës ishte kthyer mbi boshtin horizontal;
    forma e valës ishte tension pulsues DC;
     amplituda e sinjalit ishte më e vogël për shkak të rënies së tensionit në dioda.


Në ekranin e oshiloskopit u vu re gjithashtu:
Formë valore e drejtëzuar me valë të plotë;
sinjal me maja periodike pozitive;
frekuencë më e madhe në dalje 

Frekuenca në dalje rezulton:

$$
f_{out}=2f_{in}
$$

sepse ura përdor të dy gjysmëperiodat e tensionit hyrës.

**Analiza e rezultatit**

Forma e valës së shfaqur në oshiloskop konfirmon funksionimin korrekt të urës së diodave. Tensioni alternativ sinusoidal u shndërrua në tension njëkahor pulsues, ku të gjitha alternancat paraqiten pozitive.

Kjo ndodh sepse:gjatë alternances pozitive përcjell një çift diodash;
gjatë alternancës negative përcjell çifti tjetër;
drejtimi i rrymës në ngarkesë mbetet konstant.


Dalja nuk është tension DC perfekt, pasi ekziston ende valëzim (ripple), por sinjali nuk merr më vlera negative. Humbja e amplitudës vjen nga tensioni i rënies në dioda gjatë përçimit.

**Përfundime**

Eksperimenti tregoi funksionimin e saktë të urës së diodave dhe procesin e drejtëzimit të plotë të tensionit alternativ.

Nga analiza e oshiloskopit u konstatua se:
 gjysmëperiodat negative u kthyen në pozitive;
tensioni në dalje ishte pulsues DC;
ura e diodave përdori të dy gjysmëperiodat e sinjalit hyrës;
amplituda u zvogëlua për shkak të rënies së tensionit në dioda.
