**1. Qëllimi i Eksperimentit**

Qëllimi i këtij eksperimenti është studimi i sjelljes së tranzistorit bipolar NPN 2N3904 në një qark polarizimi me rezistenca në bazë, kolektor dhe emitor. Në këtë eksperiment analizohen rrymat dhe tensionet në terminalet e tranzistorit. Verifikohet ligji i ruajtjes së rrymës në tranzistor:

$$
I_E = I_B + I_C
$$

si dhe marrëdhënia ndërmjet tensionit dhe rrymës në rezistencën e emitorit sipas ligjit të Ohmit:

$$
I_E = \frac{U_E}{R_E}
$$

Gjithashtu vlerësohet përputhja ndërmjet rezultateve teorike dhe matjeve eksperimentale.

 **2. Baza Teorike**

Tranzistori bipolar NPN përbëhet nga tre terminale:
1.Baza (B)
2.Kolektori (C)
3.Emitori (E)


Në tranzistor vlen marrëdhënia:

$$
I_E = I_C + I_B
$$

Rryma e emitorit është shuma e rrymës së bazës dhe kolektorit.

Gjithashtu për rezistencën e emitorit zbatohet ligji i Ohmit:

$$
I_E = \frac{U_E}{R_E}
$$

ku:

$$
I_E = \text{rryma e emitorit}
$$

$$
U_E = \text{tensioni në emitor}
$$

$$
R_E = \text{rezistenca e emitorit}
$$

## Verifikimi i marrëdhënies:

$$

$$

$$

I_E(\text{mat}) = 0.924\,mA
$$

$$
\Delta I = 0.924 - 0.905
$$

$$
\Delta I = 0.019\,mA

$$
0.924\,mA \approx 0.905\,mA
$$

Pra marrëdhënia:

$$
I_E = I_B + I_C
$$

verifikohet afërsisht.

## Verifikimi i marrëdhënies:

$$
I_E = \frac{U_E}{R_E}

$$

I_E = \frac{0.897}{1000}

$$
I_E = 0.000897\,A

# 6. Analiza e Gabimeve

# 7. Diskutimi

Rezultatet tregojnë se tranzistori 2N3904 funksionon sipas teorisë së tranzistorëve bipolarë. Rryma e emitorit rezulton afërsisht e barabartë me shumën e rrymave të bazës dhe kolektorit, duke konfirmuar ligjin e ruajtjes së ngarkesës. Po ashtu, tensioni dhe rryma në rezistencën e emitorit ndjekin ligjin e Ohmit.

Devijimet e vogla mund të vijnë nga:

- toleranca e rezistencave
- gabime instrumentale
- rezistenca e brendshme e aparateve matëse
- temperatura e tranzistorit
- pasaktësi gjatë matjeve
