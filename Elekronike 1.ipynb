**Eksperimenti 1: Verifikimi i Ndarësit të Tensionit**

**1. Qëllimi i Eksperimentit**

Qëllimi i këtij eksperimenti është verifikimi i ligjit të ndarësit të tensionit në një qark me dy rezistenca të lidhura në seri. Analizohet shpërndarja e tensionit në funksion të vlerave të rezistencave dhe krahasohet rezultati eksperimental me atë teorik. Në veçanti, shqyrtohet rasti kur rezistencat janë pothuajse të barabarta, ku pritet që tensioni në dalje të jetë afërsisht gjysma e tensionit në hyrje.


**2. Baza Teorike**

**2.1 Modeli i Qarkut**

Qarku përbëhet nga dy rezistenca $R_1$ dhe $R_2$ të lidhura në seri me një burim tensioni $U_h$. Tensioni në dalje $U_d$ merret në skajet e rezistencës $R_2$.

**2.2 Ekuacionet Themelore**

Në një lidhje seri:

$$
I = \frac{U_h}{R_1 + R_2}
$$

Tensioni në dalje:

$$
U_d = I \cdot R_2
$$

Duke zëvendësuar:

$$
U_d = U_h \cdot \frac{R_2}{R_1 + R_2}
$$

**3.1 Montimi Praktik**

U realizua një qark me dy rezistenca në seri. Tensioni i hyrjes u aplikua në skajet e qarkut dhe tensioni i daljes u mat në $R_2$.



**4. Të Dhënat e Matura**

* $R_1 = 14.97 , k\Omega$
* $R_2 = 14.98 , k\Omega$
* $U_h = 11.57 , V$
* $U_d = 5.77 , V$


**5. Analiza e të Dhënave**

Llogaritja teorike:

$$
U_d = 11.57 \cdot \frac{14.98}{14.97 + 14.98}
$$

$$
R_1 + R_2 = 29.95 , k\Omega
$$

$$
\frac{14.98}{29.95} \approx 0.5
$$

$$
U_d^{teor} = 11.57 \cdot 0.5 = 5.785 , V
$$

Krahasimi:

* $U_d^{exp} = 5.77 , V$
* $U_d^{teor} = 5.785 , V$


**6. Analiza e Gabimeve**

Gabimi absolut:

$$
\Delta U = |5.77 - 5.785| = 0.015 , V
$$

Gabimi relativ:

$$
\varepsilon = \frac{0.015}{5.785} \cdot 100% \approx 0.26%
$$


**7. Diskutimi**

Rezultatet tregojnë se tensioni në dalje është praktikisht gjysma e tensionit në hyrje, siç parashikon teoria kur $R_1 \approx R_2$. Devijimi i vogël vjen nga toleranca e rezistencave dhe kufizimet e instrumenteve matëse. Qarku sillet sipas modelit teorik.



**8. Përfundimi**

**Eksperimenti 2: Ndikimi i Rezistencës së Ngarkesës**

**1. Qëllimi i Eksperimentit**

Qëllimi i këtij eksperimenti është të studiohet ndikimi i një rezistence të tretë (rezistencë ngarkese $R_l$) mbi tensionin në dalje të një ndarësi tensioni. Analizohet sesi ndryshon tensioni $U_d$ në varësi të madhësisë së $R_l$ dhe krahasohet me sjelljen teorike të qarkut.



**2. Baza Teorike**

Kur një rezistencë $R_l$ lidhet paralelisht me $R_2$, rezistenca ekuivalente bëhet:

$$
R_{eq} = \frac{R_2 \cdot R_l}{R_2 + R_l}
$$

Tensioni në dalje:

$$
U_d = U_h \cdot \frac{R_{eq}}{R_1 + R_{eq}}
$$

**Rastet karakteristike:**

* Kur $R_l \ll R_2 \Rightarrow R_{eq} \ll R_2$
* Kur $R_l \approx R_2 \Rightarrow R_{eq} < R_2$
* Kur $R_l \gg R_2 \Rightarrow R_{eq} \approx R_2$

**3. Ndërtimi Eksperimental**

U përdor i njëjti qark si në Eksperimentin 1 dhe u shtua një rezistencë e tretë $R_l$ paralelisht me $R_2$. U mat tensioni në dalje për vlera të ndryshme të $R_l$.


**4. Të Dhënat e Matura**

* $R_2 = 14.98 , k\Omega$
* $U_h = 11.57 , V$

Pa ngarkesë:

* $U_d = 5.77 , V$
* $\frac{U_d}{U_h} = 0.5$


**5. Analiza e të Dhënave**

🔹 Rasti 1: $R_l = 3.27 , k\Omega$

$$

$$

$$

Rezultati: Tensioni bie shumë sepse $R_l$ është i vogël.

**Rasti 2: $R_l = 330 , k\Omega$**

 **Rezultati: Tensioni është afër vlerës teorike.**

 Rezultati: Qarku sillet pothuajse si pa ngarkesë

 **6. Analiza e Gabimeve**

**Eksperimenti 3: Verifikimi i Teoremës së Thevenin-it**

**1. Qëllimi i Eksperimentit**

Qëllimi i këtij eksperimenti është verifikimi i Teoremës së Thevenin-it për një qark rezistiv. Synohet përcaktimi i tensionit ekuivalent $V_{th}$ dhe rezistencës ekuivalente $R_{th}$, si dhe krahasimi i rezultateve eksperimentale me ato teorike.


**2. Baza Teorike**

Sipas Teoremës së Thevenin-it, çdo qark linear mund të zëvendësohet me një burim tensioni $V_{th}$ dhe një rezistencë $R_{th}$ në seri.

* $V_{th}$ matet kur qarku është i hapur
* $R_{th}$ matet duke shkurtuar burimin

**Për ndarësin e tensionit:**

$$
V_{th} = U_h \cdot \frac{R_2}{R_1 + R_2}
$$

**3. Ndërtimi Eksperimental**

U ndërtua qarku me dy rezistenca në seri. Fillimisht u mat tensioni në dalje pa ngarkesë për të gjetur $V_{th}$. Më pas burimi u shkurtua për të matur $R_{th}$.


**4. Të Dhënat e Matura**

* $R_2 = 14.58 , k\Omega$
* $U_h = 11.55 , V$
* $V_{th}^{exp} = 5.73 , V$

 **5. Analiza e të Dhënave**

Llogaritja teorike:

$$
V_{th} = 11.55 \cdot \frac{14.58}{14.68 + 14.58}
$$

$$
R_1 + R_2 = 29.26 , k\Omega
$$

$$
\frac{14.58}{29.26} \approx 0.498
$$

$$
V_{th}^{teor} \approx 11.55 \cdot 0.498 = 5.75 , V
$$

**6. Analiza e Gabimeve**

Gabimi absolut:

$$

Gabimi relativ:

$$



**7. Diskutimi**

Rezultatet tregojnë një përputhje shumë të mirë midis vlerave teorike dhe atyre eksperimentale. Devijimi shumë i vogël (~0.35%) vjen nga toleranca e rezistencave dhe pasaktësitë e instrumenteve. Qarku sillet sipas modelit të Thevenin-it.



**8. Përfundimi**

Eksperimenti konfirmoi me saktësi të lartë Teoremën e Thevenin-it. Vlera teorike $5.75 , V$ dhe ajo eksperimentale $5.73 , V$ janë pothuajse identike. Devijimi shumë i vogël tregon që modeli teorik përshkruan saktë qarkun real.

**Detyre shtepie**
Të vërtetohet se fuqia maksimale P_{\text{max}} merret kur rezistenca e ngarkesës është e barabartë me rezistencën Teorema e Thevenin-it të qarkut.

$$
I = \frac{V_{th}}{R_{th} + R_L}
$$

$$
P = I^2 R_L
$$

$$
P(R_L) = \left(\frac{V_{th}}{R_{th} + R_L}\right)^2 R_L
$$

$$
P(R_L) = \frac{V_{th}^2 R_L}{(R_{th} + R_L)^2}
$$

$$
\frac{dP}{dR_L} = \frac{d}{dR_L} \left( \frac{V_{th}^2 R_L}{(R_{th} + R_L)^2} \right)
$$

$$
\frac{dP}{dR_L} = V_{th}^2 \cdot \frac{(R_{th}+R_L)^2 - 2R_L(R_{th}+R_L)}{(R_{th}+R_L)^4}
$$

$$
(R_{th}+R_L)^2 - 2R_L(R_{th}+R_L) = 0
$$

$$
(R_{th}+R_L)\big[(R_{th}+R_L) - 2R_L\big] = 0
$$

$$
(R_{th}+R_L)(R_{th} - R_L) = 0
$$

$$
R_{th} - R_L = 0
$$

$$
R_L = R_{th}
$$

Rezultatet eksperimentale janë në përputhje me teorinë, duke vërtetuar se fuqia maksimale merret kur rezistenca e ngarkesës është e barabartë me rezistencën Thevenin të qarkut.
